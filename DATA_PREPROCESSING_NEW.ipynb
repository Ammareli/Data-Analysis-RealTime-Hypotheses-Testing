```python
def detect_outlier(data):
    data = list(data)
    Q1,Q3 = np.percentile(data,[25,75])
    IQR = Q3-Q1  # IQR = Inter Quartile Range
    lower_fence = Q1-(1.5*IQR)
    high_fence = Q3 + (1.5*IQR)

    return [lower_fence,high_fence]
```

# outliers
    Hence there are outliers in both IDV and DV so for Regression Anylysis Removing Outliers

```python
def remove_outliers(data):
    normilized_data = []
    L_F,H_F = detect_outlier(data) # returns the Higher Fence and lower Fence of the column
    for i in range(len(data)):
        if data[i] <= H_F and data[i] >= L_F:
            normilized_data.append(data[i])
        else:
            continue
    return normilized_data
```

```python
conversion_speed = []
for i in range(len(copy_data)):
    conversion_speed.append(copy_data["Conversion speed"][i].days)
    
```

```python
def size_match(x,y):
    min_length = min(len(x), len(y))
    list1 = x[:min_length]
    list2 = y[:min_length]
    return list1 , list2
        
```

```python
client_age = remove_outliers(copy_data["Client_age"])
con_speed = remove_outliers(conversion_speed)
```
